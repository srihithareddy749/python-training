```python
n = int(input())
lst = list(map(int, input().split()))
S = int(input())

for i in range(n):
    for j in range(i + 1, n):
        if lst[i] + lst[j] == S:
            print((lst[i], lst[j]), end=" ")
```

```python
n = int(input())
lst = list(map(int, input().split()))

result = ()

for i in lst:
    if i % 2 == 0:
        result += (i * i,)

print(result)
```

```python
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))

t1 = tuple(list1)
t2 = tuple(list2)

temp = t1 + t1

if t2 in [temp[i:i+len(t2)] for i in range(len(t1))]:
    print(True)
else:
    print(False)
```

```python
n = int(input())
lst = list(map(int, input().split()))

t = tuple(lst)
t = tuple(set(t))   # remove duplicates

t = sorted(t)

if len(t) > 1:
    print(t[-2])
else:
    print(t[0])
```
